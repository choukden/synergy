# OneView Global Dashboard RestAPI

## Anmeldung an der Appliance

```bash
OVGDIP=10.0.20.28

sessionID=$(curl -s -X POST \
  https://${OVGDIP}/rest/login-sessions \
  -H 'Content-Type: application/json' \
  -H 'X-API-Version: 2' \
  -k \
  -d '
{
"authLoginDomain":"local",
"password":"Passw0rd",
"userName":"localUser"
}' | jq -r ".sessionID")

#sessionID=$(echo $login| awk ' { split($0,a,"sessionID\":\""); print substr(a[2],1,length(a[2])-2)}')
echo $sessionID
```

## Liste der verbundenen Appliances 

```bash
curl -s -X GET \
  https://${OVGDIP}/rest/appliances \
  -H 'Content-Type: application/json' \
  -H "auth: $sessionID" \
  -H 'x-api-version: 2' \
  -k | python -m json.tool
```

```bash
curl -s -X GET \
  https://${OVGDIP}/rest/appliances \
  -H 'Content-Type: application/json' \
  -H "auth: $sessionID" \
  -H 'x-api-version: 2' \
  -k | jq -r ".members[] | { Appliance: .applianceName, IP: .applianceLocation, Type: .model, UUID: .id }"
  
```

# OneView-Redfish

## Install toolkit
https://github.com/HewlettPackard/oneview-redfish-toolkit

```bash
pip3  install oneview-redfish-toolkit --user
```

## create configfile

```bash
cat <<'EOF' >redfish.conf
[redfish]
indent_json = True
xml_prettify = True
redfish_host = 0.0.0.0
redfish_port = 5000
authentication_mode = session

[redfish-composition]
PowerOffServerOnCompose= ForceOff
PowerOffServerOnDecompose= ForceOff

[oneview_config]
ip = 10.0.20.50

[credentials]
userName = nagios
password = Passw0rd
authLoginDomain = Local

[event_service]
DeliveryRetryAttempts = 3
DeliveryRetryIntervalSeconds = 30

[ssl]
SSLType = self-signed
SSLCertFile =
SSLKeyFile =

[ssl-cert-defaults]
countryName = US
stateOrProvinceName = California
localityName = Palo Alto
organizationName = Hewlett Packard Enterprise
#organizationalUnitName =
#commonName = SERVER_FQDN OR IP
#emailAddress = Optional sever admin email

[cherrypy_config]
server.thread_pool = 10
server.thread_pool_max = -1
server.request_queue_size = 5
EOF
```

## start oneview-redfish daeom on port localhost:5000 

```bash
oneview-redfish-toolkit --config redfish.conf 
```

## login to oneview-redfish 

```bash
REDsessionID=$(curl -i -k -s -X POST \
  -H "Content-Type:application/json" \
  https://localhost:5000/redfish/v1/SessionService/Sessions \
  -d '{"UserName": "kochstep", "Password": "Ratingen.123"}' |awk '/X-Auth/ { print $2 }')
echo $REDsessionID
```

## get list of systems 

```bash
curl -k -s -X GET \
  -H "X-Auth-Token: ${REDsessionID}" \
  https://localhost:5000/redfish/v1/Chassis/      #| jq -r ".Members"
```
